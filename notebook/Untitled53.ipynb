```python
df['Location'] = df['Location'].str.title().str.strip()
location_counts = df['Location'].value_counts()
df['Location_Order_Volume'] = df['Location'].map(location_counts)
```

```python
# Define base prices per pizza type
base_prices = {
    'Non-Veg': 10,
    'Veg': 9,
    'Cheese Burst': 11,
    'Vegan': 10,
    'Sicilian': 12,
    'BBQ Chicken': 12,
    'Margarita': 8,
    'Gluten-Free': 11,
    'Thai Chicken': 13,
    'Stuffed Crust': 12,
    'Deep Dish': 14,
    'Thin Crust': 9
}

# Size price modifiers
size_modifier = {
    'Small': 0,
    'Medium': 2,
    'Large': 4,
    'XL': 6
}

# Generate full pricing dictionary
pricing = {}
for size in size_modifier:
    for pizza_type, base_price in base_prices.items():
        pricing[(size, pizza_type)] = base_price + size_modifier[size]

# Apply pricing to the DataFrame
df['Sales ($)'] = df.apply(lambda row: pricing.get((row['Pizza Size'], row['Pizza Type']), 0), axis=1)
```

```python
Sales_by_Location = df.groupby('Location')['Sales ($)'].sum()
Sales_by_Location
```

```python
Best_Selling_Pizza = df.groupby(['Pizza Type', 'Pizza Size'])['Sales ($)'].sum()
Best_Selling_Pizza
```

```python
Sales_by_day = df.groupby(df['Order Time'].dt.date)['Sales ($)'].sum()
Sales_by_day
```

```python
df['Order Time'] = pd.to_datetime(df['Order Time'])

# Extract hour/day/month
df['Order_Hour'] = df['Order Time'].dt.hour
df['Order_Day'] = df['Order Time'].dt.day_name()
df['Order_Month'] = df['Order Time'].dt.month_name()
df['Order_Date'] = df['Order Time'].dt.date
```

```python
df['Distance Bucket'] = pd.cut(df['Distance (km)'],
                                bins=[0, 2, 5, 10, 20],
                                labels=['0–2 km', '2–5 km', '5–10 km', '10–20 km'])

# Save as CSV
df.to_csv("pizza_sales_analysis.csv", index=False)
```
